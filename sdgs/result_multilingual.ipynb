```python
import matplotlib.pyplot as plt
import numpy as np

# Generate an array of indices for the classes
classes = [str(i+1) for i in range(len(f1_micro_cso))]

# Create an array of positions for the classes
positions = np.arange(len(classes))

# Determine the common minimum and maximum values for both micro and macro F1 scores
common_min = -0.1
common_max = 0.9

# Create a single plot
fig, ax = plt.subplots(figsize=(12, 6),dpi=300)

# Plot F1 Micro scores
ax.plot(positions, f1_micro_cso, label='IndoGovBERT SC-C1 (F1 Micro)', marker='o', linestyle='-', color='g')
ax.plot(positions, f1_micro_multi, label='Multilingual BERT (F1 Micro)', marker='x', linestyle='-', color='purple')
ax.set_xlabel('Epoch')
ax.set_ylabel('F1 Score')
#ax.set_title('Comparison of F1 Micro and F1 Macro Scores')

# Create a twin y-axis for F1 Macro scores
ax2 = ax.twinx()
ax2.plot(positions, f1_macro_cso, label='IndoGovBERT SC-C1 (F1 Macro)', marker='o', linestyle='-.', color='g')
ax2.plot(positions, f1_macro_multi, label='Multilingual BERT (F1 Macro)', marker='x', linestyle='-.', color='purple')
#ax2.set_ylabel('F1 Macro Score', color='tab:red')

# Set y-axis limits for both axes
ax.set_ylim(common_min, common_max)
ax2.set_ylim(common_min, common_max)

# Set x-ticks and labels
ax.set_xticks(positions)
ax.set_xticklabels(classes)

# Show legend
lines, labels = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax.legend(lines + lines2, labels + labels2, loc='lower right')

# Adjust the layout
plt.tight_layout()
plt.grid(True)

# Show the plot
plt.show()
```
